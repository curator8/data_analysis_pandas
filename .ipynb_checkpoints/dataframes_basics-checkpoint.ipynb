### value_counts

```python
# but super useful in series 
nba_df["Position"].value_counts(normalize = True) * 100 # normalize give relative percentage
nba_df["Salary"].value_counts() # this answers what's the most common salary

```

### Dropping null values 

```python
nba_df = pd.read_csv("data/nba.csv")
nba_df.tail()
```

```python
nba_df.dropna() #drops all NaN values so it will drop rows with any NaN values 
```

```python
# any rows with all columns being absent 

nba_df.dropna(how = "all")
```

```python
# drop rows with specific NaN values present 

nba_df.dropna(subset = ["College"])
nba_df.dropna(subset = ["College", "Salary"])
```

### filling missing df values with the fillna method 

```python
nba = pd.read_csv("data/nba.csv")
nba
```

```python
nba["College"].fillna("Unknown") # we make a brand new series right here
nba["College"].fillna("Unknown", inplace = True) # this changes things permanently
```

```python
nba
```

### astype method 

```python
nba = pd.read_csv("data/nba.csv").dropna(how = "all")
# can also fill in NaN values
nba.tail()
```

```python
nba["Age"].hasnans # if a series has any NaN values conversion will fail 
```

```python
nba["Age"].astype("int") # will convert float to int 
nba["Age"].astype(int) # can pass in datatype

# can convert to one type into another by chaining astype(). 
```
